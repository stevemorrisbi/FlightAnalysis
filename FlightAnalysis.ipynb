```python
import matplotlib.pyplot as plt
```

```python
# importing flight data 
data = sns.load_dataset('flights')

# printing top 10 rows
data.head(10)
```

```python
# pivoting dataset to have years as columns

df = data.pivot(index='month', columns='year', values='passengers')
df
```

```python
# creating a draft Seaborn heatmap
ax = sns.heatmap(df)

# setting axis labels
ax.set_xlabel('Flight Year')
ax.set_ylabel('Flight Month')

# displaying the heatmap
plt.show()
```

```python
# Making improvements to the initial heatmap output

# setting the size of the heatmap to 12 inches high and 8 inches wide
plt.figure(figsize = (12,8))

# reassigning the heatmap
ax = sns.heatmap(df)

# setting the month and year tick labels to horizontal
ax.set_xticklabels(ax.get_xticklabels(), rotation = 0)
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)

# adding padding to the axis titles
ax.set_xlabel('Flight Year', labelpad = 10, rotation = 0)
ax.set_ylabel('Flight Month', labelpad = 10, rotation = 0)

# printing the heatmap
plt.show()
```

```python
# Making further amendments to the heatmap above

# setting the size of the heatmap to 12 inches high and 8 inches wide
plt.figure(figsize = (12,8))

# reassigning the heatmap - changing the colour, removing the sidebar etc
ax = sns.heatmap(df, cmap='Blues', fmt='d', annot = True, annot_kws = {'size': 10}, cbar = False, linewidths = 1, linecolor = 'white')

# setting the month and year tick labels to horizontal
ax.set_xticklabels(ax.get_xticklabels(), rotation = 0)
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)

# adding padding to the axis titles
ax.set_xlabel('Flight Year', labelpad = 10, rotation = 0)
ax.set_ylabel('Flight Month', labelpad = 10, rotation = 0)

# changing the coordinates of the X and Y titles
ax.xaxis.set_label_coords(0.5, -0.15)
ax.yaxis.set_label_coords(-0.15, 0.5)

# adding a title to the visual
ax.set_title ('Passenger Flight Traffic Figures: 1949 - 1960', fontsize = 14, y = 1.05, loc='center')

# printing the heatmap
plt.show()
```
